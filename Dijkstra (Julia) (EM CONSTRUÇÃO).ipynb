```julia
#v.1
function inputStr(mensagem)
    println(mensagem)
    return readline()
end 

function gerar_grafo()
    arquivo = open("arquivo.txt", "r");
    arquivo = split(read(arquivo, String), keepempty = true)
    
    grafo = Dict()
    i, j, contador = 1, 1, 0
    aux_no = [] 
    push!(aux_no,[])

    for linha in arquivo
        contador += 1
        
        if linha == ""
            continue
        elseif j % 3 != 0
            push!(aux_no[i], linha)
        else
            linha = parse(Float64, linha)
            push!(aux_no[i], linha)
            if contador < length(arquivo) # limitar a quantidade de array
                push!(aux_no,[])
            end
            i += 1
        end
        j += 1
    end

    tamanho = i-1 

    for i = 1:tamanho
        grafo[aux_no[i][1]] = Dict()
        grafo[aux_no[i][2]] = Dict()
    end

    for i = 1:tamanho
        grafo[aux_no[i][1]][aux_no[i][2]] = aux_no[i][3]
        grafo[aux_no[i][2]][aux_no[i][1]] = aux_no[i][3]
    end
    
    #close(arquivo);

    return grafo
end

function dijkstra(no_inicial, no_final, grafo)
    nao_visitado = grafo
    menor_distancia = Dict()
    no_anterior = Dict()
    caminho = []
    infinito = Inf

    for (no,) in nao_visitado
       menor_distancia[no] = infinito    
    end

    menor_distancia[no_inicial] = 0
    no = no_inicial

    while length(nao_visitado) > 0 
        aux = nao_visitado[no]

        for (node,peso) = aux
            if peso + menor_distancia[no] < menor_distancia[node]
                menor_distancia[node] = peso + menor_distancia[no]
                no_anterior[node] = no
            end
        end  
        
        pop!(nao_visitado,no)
        no = nothing
        
        for (node,peso) in nao_visitado
            if no == nothing
                no = node
            elseif menor_distancia[node] < menor_distancia[no]
                no = node
            end
        end
        
        if no == no_final
            break
        end
    end

    no_atual = no_final

    while no_atual != no_inicial
        pushfirst!(caminho,no_atual)
        no_atual = no_anterior[no_atual]
    end
      
    pushfirst!(caminho,no_inicial)
        
    println("Peso mínimo: $(menor_distancia[no_final])")
    println("Caminho mínimo: $caminho")
    return nothing
end

function main()
    no_inicial = "a"
    no_final = "h"
    #no_inicial = inputStr("Digite o nó inicial: ")
    #no_final = inputStr("Digite o nó final: ")
    
    grafo = gerar_grafo()

    println("")
    dijkstra(no_inicial, no_final, grafo)
end

@time main()
```
