```julia
#using Pkg; Pkg.add("Plots"); Pkg.add("Plotly");  Pkg.add("Random");
#using Random; using Plots; 
#plotly();
#theme(:juno);

#=
# Parabolóide =#
funcaoX(X,NaN) = X
funcaoY(NaN,Y) = Y
funcaoZ(X,Y) = X.^2 + Y.^2

#=
# Esfera
funcaoX(X,NaN) = X
funcaoY(NaN,Y) = Y
funcaoZ(X,Y) = X.^2 + Y.^2
=#

#=
# Toro
funcaoX(X,NaN) = X
funcaoY(NaN,Y) = Y
funcaoZ(X,Y) = X.^2 + Y.^2
=#


∂X(funcao, x_anterior, x_posterior, y, h) = (funcao(x_posterior,y) - funcao(x_anterior,y)) / (2*h)  
∂Y(funcao, x, y_anterior, y_posterior, h) = (funcao(x,y_posterior) - funcao(x,y_anterior)) / (2*h)


function primeiraFF(X, Y, h, amostraXY)
    dX, dY = [], []
    
    for k in 1:length(amostraXY)
        i, j = amostraXY[k]
        
        push!(dX, (∂X(funcaoX, X[i,j-1], X[i,j+1], Y[i,j], h),
                   ∂X(funcaoY, X[i,j-1], X[i,j+1], Y[i,j], h),
                   ∂X(funcaoZ, X[i,j-1], X[i,j+1], Y[i,j], h)))
        
        push!(dY, (∂Y(funcaoX, X[i,j], Y[i-1,j], Y[i+1,j], h),
                   ∂Y(funcaoY, X[i,j], Y[i-1,j], Y[i+1,j], h),
                   ∂Y(funcaoZ, X[i,j], Y[i-1,j], Y[i+1,j], h)))
    end

    E, F, G = [], [], []

    for i in 1:length(dX)
        push!(E, dX[i].^2 |> sum)
        push!(F, dX[i] .* dY[i] |> sum)
        push!(G, dY[i].^2 |> sum)
    end

    return E, F, G
end


#=
Função responsável por construir o domínio discretizado do parabolóide.
Retorna: X, Y e dominioXY, onde (X[i,j],Y[i,j]) é um ponto do domínio, e dominioXY guarda a posição, das matrizes X e Y, 
de todos os pontos, exceto os pontos da fronteira, isto é, quando i,j = 1 ou qtd_pontos.
=#
function dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos)
    x = [i for i in range(esquerdo_x, stop=direito_x, length=qtd_pontos)]
    y = [i for i in range(inferior_y, stop=superior_y, length=qtd_pontos)]

    X = Y = zeros(qtd_pontos,qtd_pontos)
    for i in 1:qtd_pontos, j in 1:qtd_pontos
        X[i,j] = x[j]
        Y[i,j] = y[i]
    end

    dominioXY = []
    for i in 2:qtd_pontos-1, j in 2:qtd_pontos-1
        push!(dominioXY, (i,j))
    end

    return X, Y, dominioXY
end


function amostra(dominioXY, qtd_pontos)
    amostraXY = copy(dominioXY)
    shuffle!(amostraXY)
    
    qtd_dominio = length(dominioXY)
    qtd_amostra = qtd_dominio // 2

    for i in 1:qtd_dominio - qtd_amostra
        pop!(amostraXY)
    end

    return amostraXY
end


function plotar2D(X, Y, amostra_pontos = [])
    if length(amostra_pontos) == 0
        display(scatter(X,Y))
    else
        amostraX = amostraY = []
        
        for k in 1:length(amostra_pontos)
            i, j = amostra_pontos[k]
            
            push!(amostraX, X[i,j])
            push!(amostraY, Y[i,j])
        end

        display(scatter(amostraX,amostraY))
    end
    
    return NaN
end

#=
function plotar3D(X, Y, Z, amostra_pontos = [])
    if len(amostra_pontos) == 0
        plt.figure() .add_subplot(111, projection='3d') .scatter(X, Y, Z)

        plt.show()
    else:
        amostraX, amostraY, amostraZ = [], [], []

        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]

            amostraX.append(X[i][j]) 
            amostraY.append(Y[i][j])
            amostraZ.append(Z[i][j])

        plt.figure() .add_subplot(111, projection='3d') .scatter(amostraX, amostraY, amostraZ) # gca

        plt.show()

    return NaN
end
=#

function main()
    esquerdo_x, direito_x = -10, 10
    inferior_y, superior_y = -10 , 10
    qtd_pontos = 10

    X, Y, dominioXY = dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos)
    
    amostraXY = amostra(dominioXY, qtd_pontos)
    
    #Z = funcaoZ(X, Y)
    #plotar2D(X, Y)
    #plotar3D(X, Y, Z)
    #plotar2D(X, Y, amostraXY)
    #plotar3D(X, Y, Z, amostraXY)
        
    h = X[1,2] - X[1,1]

    E, F, G = primeiraFF(X, Y, h, amostraXY)

    println(" E ")
    println(E)
    println(" F ")
    println(F)
    println(" G ")
    println(G)

    return NaN
end

main()
```
