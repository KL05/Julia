# Parabolóide

```python
import random
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

%matplotlib notebook


def funcaoZ(X, Y):
    return X**2 + Y**2

def derivadaX(x_anterior, x_posterior, y, h):
    return (funcaoZ(x_posterior,y) - funcaoZ(x_anterior,y)) / (2*h)  

def derivadaY(x, y_anterior, y_posterior, h):
    return (funcaoZ(x,y_posterior) - funcaoZ(x,y_anterior)) / (2*h)    


def primeiraFF(X, Y, h, amostra_pontos):
    dX, dY = [], []
    
    for k in range(len(amostra_pontos)):
        i, j = amostra_pontos[k]
        
        dX.append(derivadaX(X[i][j-1], X[i][j+1], Y[i][j], h))
        dY.append(derivadaY(X[i][j], Y[i-1][j], Y[i+1][j], h))

    E, F, G = [], [], []
    
    for i in range(len(dX)):
        E.append(1 + dX[i]**2)
        F.append(dX[i] * dY[i])
        G.append(1 + dY[i]**2)      
    
    return E, F, G


def dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos):
    x, y = np.linspace(esquerdo_x, direito_x, qtd_pontos), np.linspace(inferior_y, superior_y, qtd_pontos)   
    X, Y = np.meshgrid(x, y)
    
    dominioXY = []
    
    for i in range(1, qtd_pontos - 1):
        for j in range(1, qtd_pontos - 1):
            dominioXY.append((i,j))

    return X, Y, dominioXY


def amostra(X, Y, dominioXY, qtd_pontos):
    random.shuffle(dominioXY)
    
    qtd_dominio = len(dominioXY)
    qtd_amostra = qtd_dominio // 2

    if qtd_dominio > qtd_amostra:
        for i in range(qtd_dominio - qtd_amostra):
            dominioXY.pop()

    return dominioXY


def plotar2D(X, Y, amostra_pontos = []):
    if len(amostra_pontos) == 0:
        plt.scatter(X,Y)

        plt.show()
    else:
        amostraX, amostraY = [], []
        
        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]
            
            amostraX.append(X[i][j]) 
            amostraY.append(Y[i][j])

        plt.scatter(amostraX,amostraY)

        plt.show()
    
    return None


def plotar3D(X, Y, Z, amostra_pontos = []):
    if len(amostra_pontos) == 0:   
        plt.figure() .add_subplot(111, projection='3d') .scatter(X, Y, Z)

        plt.show()
    else:
        amostraX, amostraY, amostraZ = [], [], []

        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]

            amostraX.append(X[i][j]) 
            amostraY.append(Y[i][j])
            amostraZ.append(Z[i][j])

        plt.figure() .add_subplot(111, projection='3d') .scatter(amostraX, amostraY, amostraZ)

        plt.show()

    return None


def main():
    esquerdo_x, direito_x = -10, 10
    inferior_y, superior_y = -10 , 10
    qtd_pontos = 10
    X, Y, dominioXY = dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos)
    
    amostra_pontos = amostra(X, Y, dominioXY, qtd_pontos)
    
    #Z = funcaoZ(X, Y)
    #plotar2D(X, Y)
    #plotar3D(X, Y, Z)
    #plotar2D(X, Y, amostra_pontos)
    #plotar3D(X, Y, Z, amostra_pontos)
        
    h = X[0][1] - X[0][0]

    E, F, G = primeiraFF(X, Y, h, amostra_pontos)

    print(" E ")
    print(E)
    print("\n F ")
    print(F)
    print("\n G ")
    print(G)
    
    return None


main()
```

# Esfera 2.0

```python
import random
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook


def parametroX(u, v):
    return (4 * u) / (u**2 + v**2 + 4)
def parametroY(u, v):
    return (4 * v) / (u**2 + v**2 + 4)
def parametroZ(u, v):
    return (2 * (u**2 + v**2)) / (u**2 + v**2 + 4)


def derivadaX(funcao, x_anterior, x_posterior, y, h):
    return (funcao(x_posterior, y) - funcao(x_anterior, y)) / (2*h)
def derivadaY(funcao, x, y_anterior, y_posterior, h):
    return (funcao(x, y_posterior) - funcao(x, y_anterior)) / (2*h)


def primeiraFF(u, v, h, amostra_pontos):
    dX, dY = [], []
    
    for k in range(len(amostra_pontos)):
        i, j = amostra_pontos[k]
        coordenadasX, coordenadasY = [], []

        coordenadasX.append(derivadaX(parametroX, u[i][j-1], u[i][j+1], v[i][j], h))
        coordenadasX.append(derivadaX(parametroY, u[i][j-1], u[i][j+1], v[i][j], h))
        coordenadasX.append(derivadaX(parametroZ, u[i][j-1], u[i][j+1], v[i][j], h))

        dX.append(coordenadasX)
        
        coordenadasY.append(derivadaY(parametroX, u[i][j], v[i-1][j], v[i+1][j], h))
        coordenadasY.append(derivadaY(parametroX, u[i][j], v[i-1][j], v[i+1][j], h))
        coordenadasY.append(derivadaY(parametroX, u[i][j], v[i-1][j], v[i+1][j], h))
        
        dY.append(coordenadasY)

    E, F, G = [], [], []

    for i in range(len(dX)):
        E.append(dX[i][0]**2 + dX[i][1]**2 + dX[i][2]**2)
        F.append(dX[i][0]*dY[i][0] + dX[i][1]*dY[i][1] + dX[i][2]*dY[i][2])
        G.append(dY[i][0]**2 + dY[i][1]**2 + dY[i][2]**2)
    
    return E, F, G


def dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos):
    x, y = np.linspace(esquerdo_x, direito_x, qtd_pontos), np.linspace(inferior_y, superior_y, qtd_pontos)   
    u, v = np.meshgrid(x, y)

    dominio_uv = []

    for i in range(1, qtd_pontos - 1):
        for j in range(1, qtd_pontos - 1):
            dominio_uv.append((i,j))
    
    return u, v, dominio_uv


def amostra(dominio_uv, qtd_pontos):
    random.shuffle(dominio_uv)
    
    qtd_dominio = len(dominio_uv)
    qtd_amostra = qtd_dominio - (qtd_dominio // 2)
    
    for i in range(qtd_amostra):
        dominio_uv.pop()
    
    return dominio_uv


def plotar2D(u, v, amostra_pontos = []):
    if len(amostra_pontos) == 0:
        plt.scatter(u,v)

        plt.show()
    else:
        amostraX, amostraY = [], []
        
        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]
            
            amostraX.append(u[i][j]) 
            amostraY.append(v[i][j])

        plt.scatter(amostraX,amostraY)

        plt.show()
    
    return None
    
    
def plotar3D(u, v, amostra_pontos = []):
    if len(amostra_pontos) == 0:
        X = parametroX(u,v)
        Y = parametroY(u,v)
        Z = parametroZ(u,v)

        plt.figure() .add_subplot(111, projection='3d') .scatter(X, Y, Z)

        plt.show()
    else:
        amostraX, amostraY, amostraZ = [], [], []
        
        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]
            
            amostraX.append(parametroX(u[i][j], v[i][j]))
            amostraY.append(parametroY(u[i][j], v[i][j]))
            amostraZ.append(parametroZ(u[i][j], v[i][j]))

        plt.figure() .add_subplot(111, projection='3d') .scatter(amostraX, amostraY, amostraZ)

        plt.show()

    return None


def main():
    esquerdo_x, direito_x = -10, 10
    inferior_y, superior_y = -10, 10
    qtd_pontos = 10
    
    u, v, dominio_uv = dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos)

    amostra_pontos = amostra(dominio_uv, qtd_pontos)

    #plotar2D(u, v)
    #plotar3D(u, v)
    #plotar2D(u, v, amostra_pontos)
    #plotar3D(u, v, amostra_pontos)

    h = u[0][1] - u[0][0]
    
    E, F, G = primeiraFF(u, v, h, amostra_pontos)

    print(" E ")
    print(E)
    print("\n F ")
    print(F)
    print("\n G ")
    print(G)

    return None


main()
```

```python
import random
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook


def parametroX(u, v):
    return (4 + np.cos(u)) * np.sin(v)
def parametroY(u, v):
    return (4 + np.cos(u)) * np.cos(v)
def parametroZ(u, v):
    return np.sin(u)


def derivadaX(funcao, x_anterior, x_posterior, y, h):
    return (funcao(x_posterior, y) - funcao(x_anterior, y)) / (2*h)
def derivadaY(funcao, x, y_anterior, y_posterior, h):
    return (funcao(x, y_posterior) - funcao(x, y_anterior)) / (2*h)


def primeiraFF(u, v, h, amostra_pontos):
    dX, dY = [], []
    
    for k in range(len(amostra_pontos)):
        i, j = amostra_pontos[k]
        coordenadasX, coordenadasY = [], []
        
        coordenadasX.append(derivadaX(parametroX, u[i][j-1], u[i][j+1], v[i][j], h))
        coordenadasX.append(derivadaX(parametroY, u[i][j-1], u[i][j+1], v[i][j], h))
        coordenadasX.append(derivadaX(parametroZ, u[i][j-1], u[i][j+1], v[i][j], h))
        
        dX.append(coordenadasX)
        
        coordenadasY.append(derivadaY(parametroX, u[i][j], v[i-1][j], v[i+1][j], h))
        coordenadasY.append(derivadaY(parametroX, u[i][j], v[i-1][j], v[i+1][j], h))
        coordenadasY.append(derivadaY(parametroX, u[i][j], v[i-1][j], v[i+1][j], h))
        
        dY.append(coordenadasY)

    E, F, G = [], [], []

    for i in range(len(dX)):
        E.append(dX[i][0]**2 + dX[i][1]**2 + dX[i][2]**2)
        F.append(dX[i][0]*dY[i][0] + dX[i][1]*dY[i][1] + dX[i][2]*dY[i][2])
        G.append(dY[i][0]**2 + dY[i][1]**2 + dY[i][2]**2)
    
    return E, F, G


def dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos):
    x, y = np.linspace(esquerdo_x, direito_x, qtd_pontos), np.linspace(inferior_y, superior_y, qtd_pontos)   
    u, v = np.meshgrid(x, y)

    dominio_uv = []

    for i in range(1, qtd_pontos - 1):
        for j in range(1, qtd_pontos - 1):
            dominio_uv.append((i,j))
    
    return u, v, dominio_uv


def amostra(dominio_uv, qtd_pontos):
    random.shuffle(dominio_uv)
    
    qtd_dominio = len(dominio_uv)
    qtd_amostra = qtd_dominio - (qtd_dominio // 2)
    
    for i in range(qtd_amostra):
        dominio_uv.pop()
    
    return dominio_uv


def plotar2D(u, v, amostra_pontos = []):
    if len(amostra_pontos) == 0:
        plt.scatter(u,v)

        plt.show()
    else:
        amostraX, amostraY = [], []
        
        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]
            
            amostraX.append(u[i][j]) 
            amostraY.append(v[i][j])

        plt.scatter(amostraX,amostraY)

        plt.show()
    
    return None
    
    
def plotar3D(u, v, amostra_pontos = []):
    if len(amostra_pontos) == 0:
        X = parametroX(u,v)
        Y = parametroY(u,v)
        Z = parametroZ(u,v)

        plt.figure() .add_subplot(111, projection='3d') .scatter(X, Y, Z)

        plt.show()
    else:
        amostraX, amostraY, amostraZ = [], [], []
        
        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]
            
            amostraX.append(parametroX(u[i][j], v[i][j]))
            amostraY.append(parametroY(u[i][j], v[i][j]))
            amostraZ.append(parametroZ(u[i][j], v[i][j]))

        plt.figure() .add_subplot(111, projection='3d') .scatter(amostraX, amostraY, amostraZ)

        plt.show()

    return None


def main():
    esquerdo_x, direito_x = 0, 2 * np.pi
    inferior_y, superior_y = 0, 2 * np.pi
    qtd_pontos = 10

    u, v, dominio_uv = dominio(esquerdo_x, direito_x, inferior_y, superior_y, qtd_pontos)

    amostra_pontos = amostra(dominio_uv, qtd_pontos)

    #plotar2D(u, v)
    #plotar3D(u, v)
    #plotar2D(u, v, amostra_pontos)
    #plotar3D(u, v, amostra_pontos)

    h = u[0][1] - u[0][0]
    
    E, F, G = primeiraFF(u, v, h, amostra_pontos)

    print(" E ")
    print(E)
    print("\n F ")
    print(F)
    print("\n G ")
    print(G)

    return None


main()
```

```python
#Bibliotecas
############################################
import random
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
############################################



'''
Superfície Parametrizada Regular:
S(u,v) = (x(u,v),y(u,v),z(u,v)).
'''
#Parametrização do Parabolóide: S(u,v) = (u, v, u^2 + v^2)
############################
def x(u, v):
    return u

def y(u, v):
    return v

def z(u, v):
    return u**2 + v**2
############################



#Gerar domínio retangular discretizado da Superfície S
#####################################################################################################################
def dominio(esquerdo_u, direito_u, inferior_v, superior_v, qtd_pontos):
    aux1, aux2 = np.linspace(esquerdo_u, direito_u, qtd_pontos), np.linspace(inferior_v, superior_v, qtd_pontos)   
    u, v = np.meshgrid(aux1, aux2)

    #dominio_uv = np.zeros((qtd_pontos-2, qtd_pontos-2),dtype=tuple)
    dominio_uv = np.zeros((qtd_pontos-2)**2, dtype=tuple)
    
    k = 0
    
    for i in range(1, qtd_pontos - 1):
        for j in range(1, qtd_pontos - 1):
            dominio_uv[k] = (i,j)
            
            k += 1

    return u, v, dominio_uv
#####################################################################################################################



#Amostragem de pontos do domínio retangular discretizado
###################################################################
def amostra(dominio_uv, qtd_pontos):
    random.shuffle(dominio_uv)
    
    qtd_dominio = len(dominio_uv)
    
    # A quantidade de amostra será metade dos pontos do domínio
    qtd_amostra = qtd_dominio - (qtd_dominio // 2)    
        
    for i in range(qtd_amostra):
        np.delete(dominio_uv,-1)
    
    return dominio_uv
###################################################################



#Derivadas de Primeira Ordem
######################################################################
def dU(funcao, u_anterior, u_posterior, v, h):
    return (funcao(u_posterior, v) - funcao(u_anterior, v)) / (2*h)

def dV(funcao, u, v_anterior, v_posterior, h):
    return (funcao(u, v_posterior) - funcao(u, v_anterior)) / (2*h)

def derivadas(u, v, h, amostra_pontos):
    derivadaU = np.zeros(len(amostra_pontos),dtype=tuple)
    derivadaV = np.zeros(len(amostra_pontos),dtype=tuple)
    
    k = 0
    
    for coordenada in range(len(amostra_pontos)):
        i, j = amostra_pontos[coordenada]
        
        aux1 = dU(x, u[i][j-1], u[i][j+1], v[i][j], h)
        aux2 = dU(y, u[i][j-1], u[i][j+1], v[i][j], h)
        aux3 = dU(z, u[i][j-1], u[i][j+1], v[i][j], h)
        derivadaU[k] = (aux1,aux2,aux3)
        
        aux1 = dV(x, u[i][j], v[i-1][j], v[i+1][j], h)
        aux2 = dV(y, u[i][j], v[i-1][j], v[i+1][j], h)
        aux3 = dV(z, u[i][j], v[i-1][j], v[i+1][j], h)
        derivadaV[k] = (aux1,aux2,aux3)
        
        k += 1

    return derivadaU, derivadaV
######################################################################


'''
#Derivadas de Segunda Ordem
####################################################################################################################
def ddU(funcao, u_anterior, u, u_posterior, v, h):
    return (funcao(u_posterior, v) - 2*funcao(u,v) + funcao(u_anterior, v)) / (h**2)

def ddUV(funcao, u_anterior, u_posterior, v_anterior, v_posterior, h):
    return (funcao(u_posterior, v_posterior) - funcao(u_anterior, v_posterior) + funcao(u_anterior, v_anterior) + funcao(u_posterior, v_anterior)) / (4*h*h)

def ddV(funcao, u, v_anterior, v, v_posterior, h):
    return (funcao(u, v_posterior) - 2*funcao(u,v) + funcao(u, v_anterior)) / (h**2)

def derivadas2(u, v, h, amostra_pontos):
    d2U = np.zeros(len(amostra_pontos),dtype=tuple)
    d2UV = np.zeros(len(amostra_pontos),dtype=tuple)
    d2V = np.zeros(len(amostra_pontos),dtype=tuple)

    for k in range(len(amostra_pontos)):
        i, j = amostra_pontos[k]
        
        aux1 = ddU(x, u[i][j-1], u[i][j], u[i][j+1], v[i][j], h)
        aux2 = ddU(y, u[i][j-1], u[i][j], u[i][j+1], v[i][j], h)
        aux3 = ddU(z, u[i][j-1], u[i][j], u[i][j+1], v[i][j], h)
        d2U[k] = (aux1,aux2,aux3)
        
        aux1 = ddUV(x, u[i][j-1], u[i][j+1], v[i-1][j], v[i+1][j], h)
        aux2 = ddUV(y, u[i][j-1], u[i][j+1], v[i-1][j], v[i+1][j], h)
        aux3 = ddUV(z, u[i][j-1], u[i][j+1], v[i-1][j], v[i+1][j], h)
        d2UV[k] = (aux1,aux2,aux3)
        
        aux1 = ddV(x, u[i][j], v[i-1][j], v[i][j], v[i+1][j], h)
        aux2 = ddV(y, u[i][j], v[i-1][j], v[i][j], v[i+1][j], h)
        aux3 = ddV(z, u[i][j], v[i-1][j], v[i][j], v[i+1][j], h)
        d2V[k] = (aux1,aux2,aux3)

    return d2U, d2UV, d2V
####################################################################################################################
'''


#Coeficientes da Primeira Forma Fundamental
#############################################################################
def primeiraFF(derivadaU, derivadaV): 
    E = np.zeros(len(derivadaU))
    F = np.zeros(len(derivadaU))
    G = np.zeros(len(derivadaU))

    for i in range(len(derivadaU)):
        E[i] = derivadaU[i][0]**2 + derivadaU[i][1]**2 + derivadaU[i][2]**2
        F[i] = derivadaU[i][0]*derivadaV[i][0] + derivadaU[i][1]*derivadaV[i][1] + derivadaU[i][2]*derivadaV[i][2]
        G[i] = derivadaV[i][0]**2 + derivadaV[i][1]**2 + derivadaV[i][2]**2

    print(f'E \n {E} \n\n F \n {F} \n\n G \n {G}')

    return E, F, G
#############################################################################


'''
# Coeficientes da Segunda Forma Fundamental
###########################################################################################################
def segundaFF(d2X, d2XY, d2Y, normal): 
    e, f, g = [], [], []

    for i in range(len(d2X)):
        e.append(d2X[i][0]*normal[i][0] + d2X[i][1]*normal[i][1] + d2X[i][2]*normal[i][2]) # np.zeros()
        f.append(d2XY[i][0]*normal[i][0] + d2XY[i][1]*normal[i][1] + d2XY[i][2]*normal[i][2])
        g.append(d2Y[i][0]*normal[i][0] + d2Y[i][1]*normal[i][1] + d2Y[i][2]*normal[i][2])

    print(f'e \n {e} \n\n f \n {f} \n\n g \n {g}')

    return e, f, g
###########################################################################################################



#Vetor Normal
###########################################################################################################
def vetorNormal(dX, dY): #derivadaU = dX, derivadaY = dV
    produto = np.zeros(len(derivadaU),dtype=tuple)
    norma = np.zeros(len(derivadaU))
    normal = np.zeros(len(derivadaU),dtype=tuple)
    
    for i in range(len(derivadaU)):
        produto[i] = (dX[i][1]*dY[i][2] - dY[i][1]*dX[i][2], dY[i][0]*dX[i][2] - dX[i][0]*dY[i][2], dX[i][0]*dY[i][1] - dY[i][0]*dX[i][1])
    
    for i in range(len(produto)):
        norma[i] = (produto[i][0]**2 + produto[i][1]**2 + produto[i][2]**2)**(1/2)
    
    for i in range(len(produto)):
        normal[i] = (produto[i][0]/norma[i], produto[i][1]/norma[i], produto[i][2]/norma[i])
    
    return normal
###########################################################################################################



#Plano Tangente
#############################################################################
def tangente(u, v, derivadaU, derivadaV):
    tangenteX, tangenteY, tangenteZ = [], [], [] # np.zeros()

    X = x(u,v)
    Y = y(u,v)
    Z = z(u,v)

    x, y = np.linspace(-1, 7, 10), np.linspace(-1, 7, 10)
    xi, yi = np.meshgrid(x, y)

    for i in range(10): 
        for j in range (10):
            tangenteX.append(X + xi[i][j]*derivadaU[0] + yi[i][j]*derivadaV[0])
            tangenteY.append(Y + xi[i][j]*derivadaU[1] + yi[i][j]*derivadaV[1])
            tangenteZ.append(Z + xi[i][j]*derivadaU[2] + yi[i][j]*derivadaV[2])

    return tangenteX, tangenteY, tangenteZ
#############################################################################
'''


#Plotagem
########################################################################################################
def plotar2D(u, v, amostra_pontos = []):
    if len(amostra_pontos) == 0:
        plt.scatter(u,v)
        plt.show()
    else:
        amostraX, amostraY = [], []

        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]

            amostraX.append(u[i][j]) 
            amostraY.append(v[i][j])

        plt.scatter(amostraX,amostraY)
        plt.show()

    return None

def plotar3D(u, v, tangenteX, tangenteY, tangenteZ, amostra_pontos = []):
    if len(amostra_pontos) == 0:
        X = parametroX(u,v)
        Y = parametroY(u,v)
        Z = parametroZ(u,v)

        
        #Definindo o ambiente para plotar os graficos.
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')

        ax.scatter(tangenteX, tangenteY, tangenteZ)
        ax.scatter(X, Y, Z)

        plt.show()
    else:
        amostraX, amostraY, amostraZ = [], [], []

        for k in range(len(amostra_pontos)):
            i, j = amostra_pontos[k]

            amostraX.append(parametroX(u[i][j], v[i][j]))
            amostraY.append(parametroY(u[i][j], v[i][j]))
            amostraZ.append(parametroZ(u[i][j], v[i][j]))

        plt.figure() .add_subplot(111, projection='3d') .scatter(amostraX, amostraY, amostraZ)

        plt.show()

    return None
########################################################################################################





def main():
    esquerdo_u, direito_u = -10, 10
    inferior_v, superior_v = -10, 10
    qtd_pontos = 10         # quantidade de pontos nos intervalos [esquerdo_x,direito_x] e [inferior_y,superior_y]

    u, v, dominio_uv = dominio(esquerdo_u, direito_u, inferior_v, superior_v, qtd_pontos)

    amostra_pontos = amostra(dominio_uv, qtd_pontos)

    h = u[0][1] - u[0][0]

    derivadaU, derivadaV = derivadas(u, v, h, amostra_pontos)

    E, F, G = primeiraFF(derivadaU, derivadaV)

    i, j = amostra_pontos[0]
    #tangenteX, tangenteY, tangenteZ = tangente(u[i][j], v[i][j], dX[0], dY[0])

    #plotar2D(u, v)
    #plotar3D(u, v, tangenteX, tangenteY, tangenteZ)
    #plotar2D(u, v, amostra_pontos)
    #plotar3D(u, v, tangenteX, tangenteY, tangenteZ, amostra_pontos)

    return None


main()
```
