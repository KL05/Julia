```julia
# Gerar 10 valores aleatórios de 1 a 10

x = [rand(1:10) for i in 1:10]
```

```julia
# Média

media(x) = sum(x)/length(x)

media(x)
```

```julia
# Ordenando o vetor x

sort!(x)
```

```julia
# Mediana

function mediana(x)
    if length(x) % 2 == 0
        posicao_central_1 = convert(Int64, length(x)/2)
        posicao_central_2 = convert(Int64, length(x)/2 + 1)
        return (x[posicao_central_1] + x[posicao_central_2])/2
    else
        posicao_central = convert(Int64, length(x)/2 + 1)
        return x[posicao_central]
    end
end

mediana(x)
```

```julia
# Moda

#MUDA NOME DAS VARIÁVEIS E VER OUTRA FORMA MAIS SIMPLIFICADA e motrar qtd ,e caso em que não tem

function moda(x)
    moda = Dict()
    moda[x[1]] = 1
    qtd_moda = []
    push!(qtd_moda,0)
    k = 1
    
    aux_moda = []
    push!(aux_moda, x[1])
    
    for i in x
        if i in aux_moda
            l = moda[i]
            qtd_moda[l] += 1
        else
            k += 1
            push!(qtd_moda,1)
            moda[i] = k
            push!(aux_moda, i)
        end
    end
        
    qtd_maximo, posicao_maximo = findmax(qtd_moda)
    
    return aux_moda[posicao_maximo]
end

moda(x)
```

```julia
# Amplitude

amplitude(x) = x[end] - x[1]

amplitude(x)
```

```julia
# Variância

variancia(x) = [((xi - media(x))^2)/(length(x)-1) for xi in x] |> sum

variancia(x)
```

```julia
# Desvio Padrão

desvio_padrao(x) = variancia(x) |> sqrt

desvio_padrao(x)
```

```julia
# Coeficiente de Variância

coeficiente(x) = (desvio_padrao(x)/media(x)) * 100

coeficiente(x)
```

```julia
# Quartil

function quartil(x)
    q = []
    for i in 1:3
        k = div(i*(length(x)+1),4)
        push!(q, x[k] + (i*(length(x)+1)/4 - k)*(x[k+1] - x[k]))
    end
    return q
end

quartil(x)
```

```julia
# Tri-Média

function tri_media(x)
    porcentagem = 15
    excluir = div(length(x)*porcentagem,100)
    soma = [i for i in x[excluir+1:end-excluir]] |> sum
    tamanho = length(x[excluir+1:end-excluir])
    
    return soma/tamanho
end

tri_media(x)
```

```julia
# Coeficiente de Assimetria

function assimetria(x)
    soma = [((xi - media(x))/desvio_padrao(x))^3 for xi in x] |> sum

    return 1/length(x) * soma
end

assimetria(x)
```

```julia
# Curtose

function curtose(x)
    soma = [((xi - media(x))/desvio_padrao(x))^4 - 3 for xi in x] |> sum

    return 1/length(x) * soma
end

curtose(x)
```
